```python
import pandas as pd
print("Pandas library imported.")

training_years = range(2017, 2023)
training_files = [f"{year}-{year+1} Champions League Stats.xlsx" for year in training_years]

df_train_list = []
for file in training_files:
    df_train_list.append(pd.read_excel(f'/content/{file}'))

df_train = pd.concat(df_train_list, ignore_index=True)

df_test = pd.read_excel('/content/2023-2024 Champions League Stats.xlsx')

print(f"Training data loaded with {len(df_train)} rows and {len(df_train.columns)} columns.")
print(f"Test data loaded with {len(df_test)} rows and {len(df_test.columns)} columns.")

df_train.columns = df_train.iloc[1]
df_train = df_train[2:].reset_index(drop=True)
df_test.columns = df_test.iloc[1]
df_test = df_test[2:].reset_index(drop=True)

print("df_train head after header correction:\n", df_train.head())
print("df_test head after header correction:\n", df_test.head())


df_train = df_train.dropna(how='all').reset_index(drop=True)
df_test = df_test.dropna(how='all').reset_index(drop=True)

print("df_train head after dropping NaN rows:\n", df_train.head())
print("df_test head after dropping NaN rows:\n", df_test.head())

df_train = df_train.dropna(subset=['Squad']).reset_index(drop=True)
df_test = df_test.dropna(subset=['Squad']).reset_index(drop=True)

# Flatten the multi-index columns to single strings
df_train.columns = [col[1] if isinstance(col, tuple) else col for col in df_train.columns]
df_test.columns = [col[1] if isinstance(col, tuple) else col for col in df_test.columns]

# Ensure the 'Squad' column is treated as a single column name, not a multi-index tuple.
# This might happen if 'Squad' was originally the first element of a tuple, but is now a string.
# The above line should handle it, but it's good to reconfirm after flattening.
# Also, explicitly check for column existence and rename if necessary based on the instructions.
# Let's verify the columns first.

print("df_train head after more rigorous cleaning and column flattening:\n", df_train.head())
print("df_train columns after cleaning:\n", df_train.columns)
print("\n---\n")
print("df_test head after more rigorous cleaning and column flattening:\n", df_test.head())
print("df_test columns after cleaning:\n", df_test.columns)

df_train = df_train[df_train['Squad'] != 'Squad'].reset_index(drop=True)
df_test = df_test[df_test['Squad'] != 'Squad'].reset_index(drop=True)

# Columns to convert to numeric
numeric_cols = ['Eliminated', 'Eliminating', 'Score', 'Pts', 'xG', 'Value']

for col in numeric_cols:
    df_train[col] = pd.to_numeric(df_train[col], errors='coerce')
    df_test[col] = pd.to_numeric(df_test[col], errors='coerce')

# Calculate 'weighted round success'
df_train['weighted_round_success'] = (5 * df_train['Eliminated'] + 2 * df_train['Eliminating'] - df_train['Score'])
df_test['weighted_round_success'] = (5 * df_test['Eliminated'] + 2 * df_test['Eliminating'] - df_test['Score'])

# Define features
features = ['Pts', 'xG', 'Value']

# Handle missing values: drop rows with NaNs in target or features for simplicity
df_train.dropna(subset=['weighted_round_success'] + features, inplace=True)
df_test.dropna(subset=['weighted_round_success'] + features, inplace=True)

print("df_train head after final cleaning and target variable calculation:\n", df_train.head())
print("df_train info after cleaning:\n")
df_train.info()
print("\n---\n")
print("df_test head after final cleaning and target variable calculation:\n", df_test.head())
print("df_test info after cleaning:\n")
df_test.info()
```

```python
from sklearn.linear_model import LinearRegression

print("LinearRegression imported successfully.")


model = LinearRegression()

X_train = df_train[features]
y_train = df_train['weighted_round_success']

model.fit(X_train, y_train)

print("Linear regression model trained successfully.")


X_test = df_test[features]

predictions = model.predict(X_test)
df_test['predicted_weighted_round_success'] = predictions

print("Predictions for 2023-2024 season generated successfully.")
print("df_test with predictions head:\n", df_test[['Squad', 'weighted_round_success', 'predicted_weighted_round_success']].head())


import matplotlib.pyplot as plt
import seaborn as sns

print("Matplotlib and Seaborn imported successfully.")


df_test_filtered = df_test[df_test['Eliminated'] > 0].copy()

# --- Plotting ---
plt.figure(figsize=(10, 7))
sns.scatterplot(
    x='weighted_round_success',
    y='predicted_weighted_round_success',
    data=df_test_filtered,
    label='Actual vs. Predicted'
)

# Get min and max values for the diagonal line from the filtered data
min_val_filtered = min(df_test_filtered['weighted_round_success'].min(), df_test_filtered['predicted_weighted_round_success'].min())
max_val_filtered = max(df_test_filtered['weighted_round_success'].max(), df_test_filtered['predicted_weighted_round_success'].max())

# Draw a diagonal line for perfect predictions
plt.plot([min_val_filtered, max_val_filtered], [min_val_filtered, max_val_filtered], color='red', linestyle='--', label='Perfect Prediction')

plt.xlabel('Actual Weighted Round Success (Filtered Data)')
plt.ylabel('Predicted Weighted Round Success (Filtered Data)')
plt.title('Actual vs. Predicted Weighted Round Success for 2023-2024 Season (Filtered)')
plt.legend()
plt.grid(True)
plt.show()

print("Visualization of actual vs. predicted weighted round success for 2023-2024 season (filtered) displayed.")


df_test_filtered = df_test[df_test['Eliminated'] > 0].copy()

from sklearn.metrics import mean_squared_error
import numpy as np

y_true_filtered = df_test_filtered['weighted_round_success']
y_pred_filtered = df_test_filtered['predicted_weighted_round_success']

rmse = np.sqrt(mean_squared_error(y_true_filtered, y_pred_filtered))

print(f"RMSE for 2023-2024 season (excluding teams with Eliminated=0): {rmse:.2f}")
```

```python
all_season_rmse = []


full_years = range(2017, 2024)
full_files = [f"{year}-{year+1} Champions League Stats.xlsx" for year in full_years]


import numpy as np
from sklearn.metrics import mean_squared_error

for i, test_file in enumerate(full_files):
    print(f"\nProcessing {test_file} as test set...")

    # Determine training files for this iteration
    train_files = [f for j, f in enumerate(full_files) if j != i]

    df_train_list = []
    for file in train_files:
        df_train_list.append(pd.read_excel(f'/content/{file}'))
    df_train = pd.concat(df_train_list, ignore_index=True)

    df_test = pd.read_excel(f'/content/{test_file}')

    # --- Data Cleaning and Preprocessing (repeated for each iteration) ---
    # Correct headers
    df_train.columns = df_train.iloc[1]
    df_train = df_train[2:].reset_index(drop=True)
    df_test.columns = df_test.iloc[1]
    df_test = df_test[2:].reset_index(drop=True)

    # Drop rows with all NaN values
    df_train = df_train.dropna(how='all').reset_index(drop=True)
    df_test = df_test.dropna(how='all').reset_index(drop=True)

    # Drop rows where 'Squad' is 'Squad' (embedded headers) and flatten columns
    df_train = df_train[df_train['Squad'] != 'Squad'].reset_index(drop=True)
    df_test = df_test[df_test['Squad'] != 'Squad'].reset_index(drop=True)

    # Flatten columns (re-apply to ensure consistency after drops)
    # This part was tricky before, ensure it correctly flattens if a column name is a tuple
    df_train.columns = [col[1] if isinstance(col, tuple) else col for col in df_train.columns]
    df_test.columns = [col[1] if isinstance(col, tuple) else col for col in df_test.columns]

    # Ensure 'Squad' is a string column, not a tuple, for consistent filtering below
    df_train.columns = [str(col) for col in df_train.columns]
    df_test.columns = [str(col) for col in df_test.columns]

    # Convert to numeric
    numeric_cols = ['Eliminated', 'Eliminating', 'Score', 'Pts', 'xG', 'Value']
    for col in numeric_cols:
        df_train[col] = pd.to_numeric(df_train[col], errors='coerce')
        df_test[col] = pd.to_numeric(df_test[col], errors='coerce')

    # Calculate 'weighted round success'
    df_train['weighted_round_success'] = (5 * df_train['Eliminated'] + 2 * df_train['Eliminating'] - df_train['Score'])
    df_test['weighted_round_success'] = (5 * df_test['Eliminated'] + 2 * df_test['Eliminating'] - df_test['Score'])

    # Define features
    features = ['Pts', 'xG', 'Value']

    # Handle missing values: drop rows with NaNs in target or features
    df_train.dropna(subset=['weighted_round_success'] + features, inplace=True)
    df_test.dropna(subset=['weighted_round_success'] + features, inplace=True)

    if df_train.empty or df_test.empty:
        print(f"Skipping {test_file}: Training or test data is empty after cleaning.")
        continue

    # --- Model Training ---
    model = LinearRegression()
    X_train = df_train[features]
    y_train = df_train['weighted_round_success']
    model.fit(X_train, y_train)

    # --- Predictions ---
    X_test = df_test[features]
    predictions = model.predict(X_test)
    df_test['predicted_weighted_round_success'] = predictions

    # --- Filter Data and Calculate RMSE ---
    df_test_filtered = df_test[df_test['Eliminated'] > 0].copy()

    if df_test_filtered.empty:
        print(f"No eliminated teams found for RMSE calculation in {test_file}. Skipping RMSE.")
        continue

    y_true_filtered = df_test_filtered['weighted_round_success']
    y_pred_filtered = df_test_filtered['predicted_weighted_round_success']

    rmse = np.sqrt(mean_squared_error(y_true_filtered, y_pred_filtered))
    all_season_rmse.append(rmse)
    print(f"RMSE for {test_file} (filtered): {rmse:.2f}")

# --- Calculate and display average RMSE ---
if all_season_rmse:
    average_rmse = np.mean(all_season_rmse)
    print(f"\nIndividual RMSEs: {all_season_rmse}")
    print(f"Average RMSE across all seasons (filtered): {average_rmse:.2f}")
else:
    print("No RMSE values were calculated.")
```
